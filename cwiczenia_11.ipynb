## 1. Edytor a procesor tekstu. Surowy tekst.

Podczas naszych zajęć z LaTeXa, ważne jest, aby zrozumieć różnicę między **edytorem tekstu** a **procesorem tekstu**.

### Edytor tekstu
Edytor tekstu to program komputerowy służący do tworzenia i modyfikacji surowego, nieformatowanego tekstu. Przykłady edytorów tekstu to Notatnik (Windows), TextEdit (macOS) czy Vim i Emacs (Linux). Edytory tekstu są często używane przez programistów, ponieważ pozwalają na pracę z kodem źródłowym w jego najprostszej formie.

### Procesor tekstu
Procesor tekstu to zaawansowany program służący do tworzenia sformatowanych dokumentów. Oferuje on różnorodne narzędzia do formatowania tekstu, wstawiania obrazów, tabel, tworzenia list i wielu innych elementów. Przykłady procesorów tekstu to Microsoft Word, Google Docs czy Pages (macOS).

## Znaczenie surowego tekstu

Surowy tekst jest podstawą pracy z LaTeXem. W przeciwieństwie do procesorów tekstu, które ukrywają szczegóły formatowania za graficznym interfejsem użytkownika (GUI), LaTeX pozwala na precyzyjne kontrolowanie wyglądu dokumentu poprzez pisanie i kompilowanie kodu.

### Dlaczego surowy tekst jest ważny:

1. **Uniwersalność**: Surowy tekst może być czytany i edytowany na każdym komputerze bez potrzeby specjalistycznego oprogramowania.
2. **Kontrola**: Daje pełną kontrolę nad formatowaniem dokumentu, co jest szczególnie ważne w pracach naukowych i technicznych.
3. **Współpraca**: Ułatwia współpracę i kontrolę wersji, ponieważ zmiany w tekście są łatwe do śledzenia.
4. **Stabilność**: Formaty surowego tekstu są mniej podatne na błędy i niezgodności między różnymi wersjami oprogramowania.

W kolejnych częściach zajęć nauczymy się, jak wykorzystać LaTeX do tworzenia profesjonalnie wyglądających dokumentów, korzystając z  surowego tekstu.



## 2. Krótka historia i główne idee


 **Donald Ervin Knuth** (ur. 10 stycznia 1938 r. w Milwaukee) – amerykański
 matematyk, informatyk, emerytowany profesor na katedrze informatyki
 Uniwersytetu Stanforda. Jeden z pionierów informatyki – jest
 najbardziej znany z wielotomowego dzieła Sztuka programowania (The Art
 of Computer Programming), uznawanego za najbardziej dogłębne, chociaż
 na razie niedokończone opracowanie na temat analizy algorytmów.

Donald Knuth rozpoczął prace nad TeXem w późnych latach 70. XX wieku, kiedy to nie był zadowolony z jakości składu swojej książki "The Art of Computer Programming". W 1978 roku opublikował pierwszą wersję TeXa, systemu składu tekstu, który pozwalał na tworzenie wysokiej jakości dokumentów z precyzyjnym formatowaniem matematycznym.

TeX był rewolucyjny, ponieważ oferował bardzo wysoki poziom kontroli nad składem tekstu, co było szczególnie ważne w środowisku akademickim i naukowym, gdzie skomplikowane wzory matematyczne, tabele i bibliografie były trudne do złożenia przy użyciu tradycyjnych maszyn do pisania czy wczesnych procesorów tekstu.

W latach 80. **Leslie Lamport**, informatyk pracujący w SRI International, zaczął pracę nad zestawem makr dla TeXa, które miały ułatwić korzystanie z systemu. Jego prace doprowadziły do stworzenia LaTeXa, który został opublikowany w 1985 roku. LaTeX budował na fundamencie TeXa, dodając zestaw predefiniowanych szablonów i makr, które umożliwiały łatwiejsze formatowanie dokumentów i zarządzanie strukturą, taką jak sekcje, rozdziały, spisy treści, przypisy i bibliografie.

LaTeX szybko zyskał popularność w środowiskach akademickich, ponieważ znacznie upraszczał proces tworzenia skomplikowanych dokumentów, zwłaszcza tych zawierających matematykę. W kolejnych latach system był rozwijany i ulepszany przez społeczność użytkowników i programistów. W 1994 roku została wydana wersja LaTeX2e, która jest do dziś standardem i najczęściej używaną wersją LaTeXa.

LaTeX jest obecnie używany na całym świecie i ma silną społeczność użytkowników, którzy tworzą i udostępniają nowe pakiety rozszerzające jego możliwości. Jest to oprogramowanie typu open-source, dostępne na różne platformy systemowe, i jest uważany za standard w składzie publikacji naukowych, zwłaszcza w dziedzinach matematyki, informatyki, fizyki i inżynierii.


### Zalety programu 

- Wszystkie zalety plików tekstowych

- Program tworzy plik pdf o wysokiej jakości (na przykład
   automatycznie łamane są wiersze), wzory matematyczne łądnie
   wyglądają

- Wsparcie dla mniej standardowych zastosowań (na przykłąd wykonywania
   rysunków, wstawiania wzorów chemicznych, rozgrywek szachowych etc.)

- Program jest darmowy

- Możliwa integracja z oprogramowaniem i automatyzacja pracy (na
   przykład generowanie plików pdf na serwerze)

- Możliwość tworzenia z jednego pliku dokumentów wielu typów (na
   przykład raportów, prezentacji etc)


### Praca z LaTeXem

Idea pracy programu jest następująca:

- tekstowy plik źródłowy (najczęściej z roszerzeniem .tex) kompilowany (określenie *kompilowany* nie jest tu do końca precyzyjne)
   jest za pomocą programu (na przykład TeX, LaTeX lub LuaTeX),

- treść dokumentu oddzielona jest od jej formatowania (w odróżnieniu
   od programów typu **WYSIWYG** (ang. what you see is what you
   get). Działa to trochę na zasadzie pliku stylu css w przypadku stron
   html,

- przy kompilacji tworzone są pliki pomocnicze, które mogą być
   przetwarzane przez inne programy (na przykład przez program
   przygotowujący skorowidz),

- jeśli to konieczne plik jest kompilowany ponownie i w wyniku
   powstaje plik pdf (najczęściej).

W naszych laboratoriach dostępne są różne edytory ułatwiające tworzenie plików latexowych: TeXStudio, TeXWorks, TeXnicCenter, TeXMaker. Niewątpliwą zaletą tych edytorów jest kolorowanie składni, skróty, szybkie tworzenie preambuł etc. Pliki można także edytować w systemowym notatniku i kompilować z linii komend.

 Schemat prostego (minimalna zawartość pakietów) dokumentu napisanego w
 języku polskim (sugeruję wpisywać "ręcznie"). Pierwsza część dokumentu  nosi nazwę **preambuły**. W części tej definiujemy klasę dokumentu (np. article, beamer, book, itd.) oraz dołączamy niezbędne pakiety oraz wprowadzamy inne istotne ustawienia. Następnie następuje **część główna** dokumentu. Prosty schemat preambuły:

    \documentclass[a4paper,11pt]{article}  %%preambuła start

    \usepackage[utf8]{inputenc}
    \usepackage{polski}

    \title{Pierwszy dokument}
    \author{Autor Dokumentu}             %%preambuła stop

    \begin{document}                     %%część główna start

    \maketitle
    
     Dowolny tekst w języku polskim.

    \end{document}                     %% część główna stop
    
## 5. Inne informacje.

Ważnym elementem w pracy z LaTeXem jest znajomość różnych pakietów pozwalających na korzystanie z dodatkowych funkcjonalności. 

Zasada wpisywania komend jest następująca: `\' + nazwa komendy, argumenty obowiązkowe w nawiasie klamrowym i argument nieobowiązkowe w nawiasie kwadratowym. Mamy również tzw. środowiska, które mają postać:

```latex
\begin{nazwa_środowiska}

\end{nazwa_środowiska}
```


Tworzenie kursywy - `\emph`, pogrubienia - `\textbf`,
nowe akapity - wolna linia, zdanie od nowej linii (`\\'), centrowanie fragmentu tekstu
(środowiska center) oraz skład do prawej strony (środowisko flushright).



Do wstawiania komentarzy używamy symbolu `%` (chodzi o wstawianie komentarza).

## 6. LaTeX Mini-Cheat Sheet.

### Formatowanie tekstu

**Pogrubienie tekstu:**  
`\textbf{tekst}`

**Kursywa:**  
`\textit{tekst}`

**Podkreślenie:**  
`\underline{tekst}`

**Zmiana rozmiaru czcionki:**  
`\tiny{tekst}`  
`\scriptsize{tekst}`  
`\footnotesize{tekst}`  
`\small{tekst}`  
`\normalsize{tekst}`  
`\large{tekst}`  
`\Large{tekst}`  
`\LARGE{tekst}`  
`\huge{tekst}`  
`\Huge{tekst}`

**Wyśrodkowanie tekstu:**  
`\begin{center}`  
`Tekst do wyśrodkowania`  
`\end{center}`

**Wyrównanie do lewej:**  
`\begin{flushleft}`  
`Tekst do wyrównania do lewej`  
`\end{flushleft}`

**Wyrównanie do prawej:**  
`\begin{flushright}`  
`Tekst do wyrównania do prawej`  
`\end{flushright}`

### Podział dokumentu 


```latex
\section{Sekcja}
Treść sekcji.

\subsection{Podsekcja}
Treść podsekcji.

\subsubsection{Podpodsekcja}
Treść podpodsekcji.

\paragraph{Paragraf}
Treść paragrafu.

\subparagraph{Podparagraf}
Treść podparagrafu.
```

### Listy

W LaTeXu możemy tworzyć listy numerowane oraz nienumerowane. Poniżej znajduje się przykład tworzenia obu typów list.

#### Lista nienumerowana

```latex
\begin{itemize}
  \item Pierwszy element listy
  \item Drugi element listy
  \item Trzeci element listy
\end{itemize}
```

#### Lista numerowana

```latex
\begin{enumerate}
  \item Pierwszy element listy
  \item Drugi element listy
  \item Trzeci element listy
\end{enumerate}
```

#### Lista opisowa

```latex
\begin{description}
  \item[Nazwa] Opis
  \item[Inna Nazwa] Inny opis
\end{description}
```


### Definiowanie nowych twierdzeń z `newtheorem`

```latex
\documentclass{article}
\usepackage[utf8]{inputenc}

% Definiowanie nowych środowisk dla twierdzeń
\newtheorem{theorem}{Twierdzenie}
\newtheorem{lemma}[theorem]{Lemat}
\newtheorem{corollary}{Wniosek}[theorem]
\newtheorem{definition}{Definicja}

\begin{document}

% Użycie nowo zdefiniowanego środowiska
\begin{theorem}
Tutaj wpisz treść twierdzenia.
\end{theorem}

\begin{lemma}
Tutaj wpisz treść lematu, który będzie miał wspólny licznik z twierdzeniami.
\end{lemma}

\begin{corollary}
Tutaj wpisz treść wniosku, który będzie miał podlicznik związany z twierdzeniami.
\end{corollary}

\begin{definition}
Tutaj wpisz treść definicji.
\end{definition}

\end{document}
```

### Etykietowanie i cytowanie formuł

```latex
\documentclass{article}
\usepackage[utf8]{inputenc}
\usepackage{amsmath}

\begin{document}

% Etykietowanie formuły
\begin{equation}\label{eq:example}
E = mc^2
\end{equation}

% Cytowanie formuły
Równanie \ref{eq:example} jest znane jako równanie Einsteina.

\end{document}
```

### Tworzenie bibliografii

```latex
\documentclass{article}
\usepackage[utf8]{inputenc}

\begin{document}

% Cytowanie w tekście
Jedną z najważniejszych prac w dziedzinie fizyki jest praca Einsteina \cite{einstein}.

% Bibliografia
\begin{thebibliography}{9}

\bibitem{einstein}
A. Einstein,
\textit{Zur Elektrodynamik bewegter Körper},
Annalen der Physik,
1905.

\end{thebibliography}

\end{document}
```


### Napisy matematyczne

**Tryb matematyczny inline (w linii tekstu):**  
`$wyrażenie matematyczne$`

**Tryb matematyczny wyświetlany (w osobnej linii):**  
`$$wyrażenie matematyczne$$`  
lub  
`\[wyrażenie matematyczne\]`

**Indeks dolny i górny:**  
`$x_i^2$` - x z indeksem dolnym i i kwadratem jako indeksem górnym

**Ułamek:**  
`\frac{licznik}{mianownik}`

**Pierwiastek:**  
`\sqrt{x}`  
`\sqrt[n]{x}` - pierwiastek n-tego stopnia z x

**Suma:**  
`\sum_{i=1}^{n} x_i` - suma x_i od i=1 do n

**Iloczyn:**  
`\prod_{i=1}^{n} x_i` - iloczyn x_i od i=1 do n

**Integral:**  
`\int_{a}^{b} f(x) \, dx` - całka funkcji f(x) od a do b

**Granica:**  
`\lim_{x \to \infty} f(x)` - granica funkcji f(x) gdy x dąży do nieskończoności

**Równania i wyrównywanie:**  
`\begin{align}`  
`x &= y + z \\`  
`y &= z - x`  
`\end{align}`

**Macierze:**  
`\begin{matrix}`  
`a & b \\`  
`c & d`  
`\end{matrix}`

**Równania z przypadkami:**  
`\begin{cases}`  
`x, & \text{jeśli } y > 0 \\`  
`-x, & \text{jeśli } y \leq 0`  
`\end{cases}`

**Symbol nieskończoności:**  
`\infty`

**Greckie litery:**  
`\alpha` - α, `\beta` - β, `\gamma` - γ, itd.

**Operator nabla:**  
`\nabla`

**Różne symbole matematyczne:**  
`\times` - znak mnożenia  
`\div` - znak dzielenia  
`\pm` - plus-minus  
`\cdot` - kropka jako znak mnożenia  
`\neq` - znak nierówności  
`\approx` - znak przybliżenia  
`\equiv` - znak równoważności  
`\leq` - mniejsze bądź równe  
`\geq` - większe bądź równe  
`\subset` - podzbiór  
`\supset` - nadzbiór  
`\in` - należy do  
`\notin` - nie należy do  
`\forall` - dla każdego  
`\exists` - istnieje  
`\nexists` - nie istnieje  
`\emptyset` - zbiór pusty  
`\land` - koniunkcja (i)  
`\lor` - alternatywa (lub)  
`\neg` - negacja  
`\to` - implikacja  
`\gets` - odwrotna implikacja  
`\iff` - równoważność (wtedy i tylko wtedy, gdy)
