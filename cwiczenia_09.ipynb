## Wolfram Alpha

### Informacje wstępne

<a href="https://www.wolframalpha.com/">**Wolfram Alpha**</a> - wyszukiwarka wiedzy typu Question Answering stworzona przez Wolphram Alpha LLC (twórcy programu Mathematica). Pierwotnie strona zawierała oprogramowanie matematyczne, obecnie jest wielodziedzinowa. Dzięki zaawansowanym technologiom NLP interpretuje zapytania w języku naturalnym.

1. Skupimy się w dużej mierze na matematycznych możliwościach *Wolfram Alpha*. Oczywiście można wpisując odpowiednie polecenia w języku angielskim w języku naturalnym uzyskać satysfakcjonujący wynik. Można posłużyć się także opcją **math input**, która działa podobnie jak wpisywanie matematycznych wyrażeń w *MSWord*. Najlepszy efekt dają odpowiednie **kwerendy**.

2. Opis poszczególnych poleceń.

    - Rysowanie krzywych i wykresów funkcji.
      >> **plot** lub **graph**  rysuje wykres podanej funkcji (jednej lub dwóch zmiennych);
      >> jeśli po wzorze dodamy **from x=x_0 to n** otrzymamy wykres funkcji w zadanym przedziale;
      >> możemy zastosować polecenie **parametric plot**, aby narysować krzywą w postaci parametrycznej, jeśli w poleceniu dopiszemy **3d** mamy wykres w przestrzeni;
      >> **polar plot** - wykres we współrzędnych biegunowych;
      >> **tangent line to** (równanie) **at x=...** rysuje styczną do wykresu funkcji (krzywej) w danym punkcie.

	- Całkowanie, różniczkowanie, obliczanie granic, szeregi liczbowe i funkcyjne.

    	>> **integrate** [wzór funkcji] **dx** - wyznaczenie całki nieoznaczonej;
	    >> **integrate**[wzór funkcji] **dx from x=a to b** - wyznaczenie całki oznaczonej, w tym, biorąc za b nieskończoność, całki niewłaściwej;
	    >> całki wielokrotne otrzymamy dopisująć **dy**, **dz** itd. 

	   >> **derivative of** [wzór funkcji] - wyznaczenie pochodnej funkcji;
	   >> alternatywnie możemy użyć zapisu **d/dx**[wzór funkcji] do liczenia pochodnych funkcji jednej i wielu zmiennych
	   >> **derivative of** [wzór funkcji] **at x=a** - wyznaczenie pochodnej funkcji w punkcie;
	   >> **second derivative of** [wzór funkcji] - wyznaczenie drugiej pochodnej funkcji;
	   >>**minimize/maximize** [wzór funkcji lub ciągu] **over [domain]** - wyznaczanie minimów/maksimów funkcji w danym obszarze;
	   
	   >> **lim** [wzór funkcji lub ciągu] **as x->x_0** - wyznaczanie granicy funkcji lub ciągu;
       >> jeśli granica ma być w nieskończoności wpisujemy **infinity**, jeśli jednostronna - znak za liczbą;
       
       >>**Sum(a_n, (n,n_0,n_k))** - obliczanie sum częściowych lub sum szeregów liczbowych oraz funkcyjnych wraz z informacją o (obszarze) zbieżności lub rozbieżności;
       >> **taylor series f(x) at x=x_0 order n** - rozwinięcia danej funkcji w szereg Taylora w danym punkcie;
       
    - Rachunek macierzowy, równania, układy równań, rozkład wielomianów.
    
    	>> **{{a_11,a_12,...,a_1n},{a_21,a_22,...,a_2n},...{a_m1,a_m2,...,a_mn}}** - w ten sposób tworzymy macierze, 
    	>> automatycznie po wciśnięciu ENTER lub kliknięciu =, pojawia się sporo informacji szczególnie dla macierzy kwadratowych - wartość wyznacznika, wielomian charakterystyczny, macierzodwrotna, diagonalizacja, wartości i wektory własne; 
    	>>działania na macierzach - tak jak działania arytmetyczne;
    	>> **factor**[wzór wielomianu] - rozkład wielomianu (zmiennej rzeczywistej i zespolonej);
    	>> tworzenie i rozwiązywanie równań i układów równań - poszczególne równania oddzielamy słowem **and** można użyć polecenia **solve**;
        
     - Elementy statystyki.
    
    	>> Aby otrzymać podstawowe statystyki opisowe wystarczy w klamrowym nawiasie (lub) wpisać po przecinku interesujące nas dane. Można użyć także komend **mean**, **var*, **std**. Polecenie **linear fit** wyszuka prostą regresji do danego zestawu danych
    	 
    >Czasem wpisanie mniej precyzyjnej kwerendy powoduje pojawienie się większej ilości informacji np. wpisanie samego wzoru funkcji spowoduje, że program wyświetli jej wykres, miejsca zerowe, informacje o monotoniczności itp.

### Prezentacja możliwości Colaba

Wprowadź kod podany niżej, aby zbudować model klasyfikacyjny dla popularnego zbioru `iris`. Na podstawie trzech wartości odpowiednich parametrów za pomocą algorytmu K-Nearest Neighbours zbudujesz (za pomocą poniższego kodu) model klasyfikacyjny.

1. **Uruchomienie nowego notatnika w Google Colab:**
   - Przejdź na stronę https://colab.research.google.com/ i zaloguj się na swoje konto Google.
   - Kliknij "New Notebook" (Nowy notatnik), aby utworzyć nowy notatnik.
   

2. **Importowanie bibliotek:**
   ```python
   # Importowanie potrzebnych bibliotek
   import pandas as pd
   import seaborn as sns
   from sklearn import datasets
   from sklearn.model_selection import train_test_split
   from sklearn.preprocessing import StandardScaler
   from sklearn.neighbors import KNeighborsClassifier
   from sklearn.metrics import classification_report, confusion_matrix
   ```

3. **Zaimportowanie danych ze zbioru iris:**
   ```python
   # Wczytanie danych
   iris = datasets.load_iris()
   iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
   iris_df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
   ```

4. **Ilustracja danych za pomocą biblioteki pandas:**
   ```python
   # Wyświetlenie pierwszych pięciu wierszy DataFrame
   print(iris_df.head())
   ```

5. **Wizualizacja danych za pomocą seaborn:**
   ```python
   # Wizualizacja danych
   sns.pairplot(iris_df, hue='species')
   ```

6. **Przygotowanie danych do modelowania:**
   ```python
   # Podział danych na cechy i etykiety
   X = iris_df.iloc[:, :-1].values
   y = iris.target

   # Podział danych na zbiór treningowy i testowy
   X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

   # Normalizacja cech
   scaler = StandardScaler()
   scaler.fit(X_train)

   X_train = scaler.transform(X_train)
   X_test = scaler.transform(X_test)
   ```

7. **Wybór i trenowanie algorytmu klasyfikacyjnego (np. k-NN):**
   ```python
   # Utworzenie klasyfikatora k-NN
   classifier = KNeighborsClassifier(n_neighbors=5)

   # Trenowanie klasyfikatora
   classifier.fit(X_train, y_train)
   ```

8. **Przewidywanie i ocena modelu:**
   ```python
   # Przewidywanie na zbiorze testowym
   y_pred = classifier.predict(X_test)

   # Ocena modelu
   print(confusion_matrix(y_test, y_pred))
   print(classification_report(y_test, y_pred))
   ```
