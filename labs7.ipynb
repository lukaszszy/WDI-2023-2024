### Informacje ogólne o HTML

**HTML** to skrót od *HyperText Markup Language*, który jest standardowym językiem do tworzenia stron internetowych. HTML definiuje strukturę i układ strony internetowej za pomocą różnych elementów tego jeżyka. Podobnie jak w przypadku XML do tworzenia kolejnych elementów używamy znaczników HTML (tzw. tagów) i atrybutów. 

### Podstawowe znaczniki HTML

Jest wiele znaczników HTML, ale podstawowe z nich to:

1. $\text{<!DOCTYPE html>}$ -  deklaruje typ dokumentu i wersję HTML.
2. $\text{<html>}$ -  określa początek i koniec dokumentu HTML.
3. $\text{<head>}$ -  zawiera meta informacje, takie jak tytuł strony.
4. $\text{<title>}$ -  określa tytuł strony.
5. $\text{<body>}$ - zawiera główną zawartość strony.
6. $\text{<h(1-6)>}$ - określają nagłówki.
7. $\text{<p>}$ - definiuje paragraf.
8. $\text{<br>}$ -  służy do wstawiania pojedynczego line break.
9. $\text{<hr>}$ - tworzy poziomatą linię, która działa jako separator treści.
10. $\text{<a>}$ - znacznik linku, który jest używany do definiowania hiperłącza.
11. $\text{<img>}$ - służy do umieszczania obrazów na stronie internetowej.
12. $\text{<div>}$ - jest to blokowy poziom elementu podziału sekcji, który jest używany do organizowania danych i bloków kodu.
13. $\text{<span>}$ - jest to znacznik, który jest używany do grupowania lub aplikowania stylów do tekstu wewnątrz elementów blokowych.
14. $\text{<ul>}$ i $\text{<li>}$ - znaczniki te są używane do tworzenia list. $\text{<ul>}$ jest używany do tworzenia listy, a $\text{<li>}$ jest używany do tworzenia elementu listy.
15. $\text{<table>}$ - jest używany do tworzenia tabeli. Kolejne znaczniki jak $\text{<tr>}$ (table row), $\text{<th>}$ (table header) i $\text{<td>}$ (table data) są używane do tworzenia wierszy, nagłówków i danych tabeli odpowiednio.

Ważnym elementem każdego języka są komentarze. W języku HTML mają one postać:

$\text{<!-- This is a comment -->}$

Więcej szczegółów można znaleźć w podlinkowanym wyżej <a href="https://www.w3schools.com/html/default.asp">kursie HTML.</a>

### Zasady stosowania CSS

**CSS (Cascading Style Sheets)** to język, który służy do stylizacji stron internetowych. 

Mamy trzy sposoby dodawania CSS:
- wewnątrz linii (inline),
- wewnętrzny arkusz (iternal),
- zewnętrzny arkusz (external).

#### Stylowanie wewnątrzliniowe

Przykład kodu z wewnątrzliniowymi stylami:

$\text{<p style="color:red;”>Czerwony tekst</p>}$

W powyższym przypadku, tekst wewnątrz znacznika będzie miał kolor czerwony.

### Porównanie metod dodawania CSS
Zewnętrzne arkusze stylów są zalecaną metodą, ponieważ ułatwiają zarządzanie kodem i utrzymanie strony.
Wewnętrzne arkusze stylów mogą być przydatne, gdy stylów jest bardzo mało lub gdy chcemy dodać specyficzne style tylko dla jednej strony.
Stylowanie wewnątrzliniowe jest niezalecane ze względu na trudności w zarządzaniu kodem i utrzymaniu strony.

### Selektory

**Selektory CSS** to wzorce używane do wybrania elementów HTML, które chcemy sformatować. Innymi słowy selektory, służą do "wybierania" elementów na stronie, do których chcemy zastosować określone style CSS.

Możemy je zastosować na wiele różnych sposobów. Oto kilka przykładów:


- **Selektor elementu (element selector)** - definiuje styl konkretnych elementów HTML. Sposób definiowania:

html-element{

property: value;

}

Przykład:

h2{

color: red;

}

spowoduje, że tekst we wszystkich elementach h2 wyświetli się na czerwono.


- **Selektor klasy (class selector)** - w kodzie HTML należy dodać atrybut *class* do wybranych elementów, dla których definiujemy styl. Ustaw wartość atrybutu na "class-name". W pliku CSS styl dla wskazanej klasy definiowany jest następująco:

.class-name{

property:value;

}

W ten sposób każdy element HTML mający atrybut class ustawiony na "class-name" ma zdefiniowany styl w podany wyżej sposób.


- **Selektor Id (selector id)** - działa  dosyć podobnie jak Class Selector z tym, że *id* stosujemy tylko dla unikalnego elementu. Sformatowane zostaną tylko elementy o wskazanym atrybucie id o podanej wartości.

#id-name{

property:value;

}

 

- **Selektor atrybutu (attribute selector)** -  definiuje styl dla tych elementów HTML  z podanym atrybutem (ewentualnie określoną wartością).

html_el[attribute]{

property: value

}

html_el[attribute="value"]{

property: value

}

Przy różnych wartościach atrybutu możemy w ten sposób zdefiniować inny styl.

- **Selektor uniwersalny (universal selector)** "Select all" - definiuje styl dla wszystkich elementów.

*{

property: value;

}
