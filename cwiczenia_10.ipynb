## Wiadomości wstępne.

**Microsoft Excel** jest arkuszem kalkulacyjnym, czyli programem, pozwalającym na automatyczną obróbkę danych, przeważnie liczbowych, przedstawionych 
w formie tabeli. Innymi znanymi tego typu programami są *OpenOffice Calc* (bezpłatny) czy *Quattro Pro*. 

Dane w Excelu przechowywane są w komórkach, które uporządkowane są podobnie jak współrzędne kartezjańskie czy elementy macierzy. Wielka litera oznacza numer kolumny, 
zaś liczba numer wiersza. Wiersze czy kolumny można rozszerzać lub zwężać przesuwając odpowiednio kursorem. Przemieszczać po arkuszu możemy się strzałkami, 
kursorem, `ENTER` przenosi nas do kolejnego wiersza, `TAB` do kolejnej komórki. Przy dużych zestawach danych warto pamiętać, że skrót `CTRL + UP/DOWN/LEFT/RIGHT`, 
przenosi do początku lub końca wiersza/kolumny,  w której się aktualnie znajdujemy. 
Warto przypomnieć o możliwości "zautomatyzowania" wprowadzania danych. Po wybraniu konkretnej komórki w prawym dolnym rogu znajduje się kwadracik. 
Gdy najeżdżamy na niego zmienia się kształt kursora. Jeśli "przeciągniemy kwadracik"  przy wciśniętym lewym przycisku myszy kopiujemy zawartość zaznaczonej
komórki do kolejnych komórek wiersza/kolumny. Podobnie możemy skopiować sekwencje kilku komórek. Jeśli Excel rozpoznaje sekwencje, to może też dopisać 
kolejne jej elementy, np. wpisując 1 i 2 w kolejnych wierszach tej samej kolumny i przeciągając zaznaczone komórki na kolejne wiersze otrzymujemy ciąg 
kolejnych liczb. Klikając w >>Opcje Autowypełniania<< możemy sami dostosować tę funkcję w zależności do naszych potrzeb.

Zawartość komórek możemy formatować w podobny sposób jak robi się to w Wordzie (justowanie, wyrównywanie pionowe, zmiana czcionki, podkreślanie, pogrubianie, etc. - wszystkie
opcje zawarte w >>Narzędziach głównych<<). 
Chociaż dane wprowadzane do Excela zorganizowane są tak jak w tabeli, to jednak przy wydruku niewidoczne są pionowe i poziome linie. 
Aby się one pojawiły należy zaznaczyć komórki z odpowiedniego zakresu,
następnie wybrać >>Obramowania<< i wybrać, którąś z interesujących nas opcji (np. wszystkie krawędzie). 
Możemy także zaznaczyć dwie lub więcej sąsiednich komórek i wybierając
opcję >>Scal i wyśrodkuj<<, która pozwoli na połączenie wybranych komórek (scalone komórki w analogiczny sposób możemy rozdzielić).

## Formuły. Adresowanie względne i bezwzględne. 

### Adresowanie

Formuły są jednym z tych elementów, które stanowią o dużej funkcjonalności Excela. Ich użycie daje nam szereg możliwości obróbki danych. 
Przeanalizujmy to na prostym przykładzie.  Utwórzmy prostą "tabelkę" składającą się z czterech kolumn (tabelkę utworzymy w lewym górnym rogu - do takiej tabelki będą odnosić się formuły). 
Scalmy cztery górne komórki pierwszego wiersza  i utwórzmy napis (tytuł tabelki) "Zakupy".  W pierwszej kolumnie kolejnego wiersza wpiszmy "artykuł", 
następnie w kolejnych: "Cena", "Ilość" oraz  "Za artykuł". 
Uzupełnijmy trzy-cztery wiersze trzech pierwszych kolumn (artykuł, cena, ilość) odpowiednimi danymi. Będziemy chcieli obliczyć koszt zakupów. 
Wykorzystamy do tego celu odpowiednią formułę. W tym celu, 
W pierwszej komórce kolumny "za artykuł" tj. D3, wpiszmy  **znak równości**, a następnie $B3*C3$. Po wciśnięciu klawisza Enter pojawi nam 
się odpowiedni wynik będący iloczynem zawartości kolumn B2 i C2,  o ile zawierają one odpowiednie dane. 
Następnie przeciągamy komórkę D3  do końca kolumny w tabeli. Ta operacja sprawi, że w komórkach od D4 do D6 pojawią się odpowiednio formuły: $B4*C4$, $B5*C5$, $B6*C6$. 
Excel wypełnił więc w analogiczny sposób komórki D4 - D6 jako iloczyn stojących obok każdej z nich komórek sąsiednich. Jest to przykład tzw. **adresowania względnego**.
Inną formą adresowania jest **adresowanie bezwzględne**, czyli takie, w którym odwołujemy się do zawartości konkretnej (nie zmieniającej położenia) komórki. 
Wtedy przy obu "współrzędnych" komórki musimy postawić znak $\$$.

Dla ilustracji adresowania bezwzględnego dodajmy kolejną kolumnę do utworzonej wyżej tabelki, w której przechowywać będziemy cenę w Euro. Utwórzmy nową komórkę, w której przechowywać będziemy kurs waluty. 
Następnie adresując bezwzględnie (przed wstawieniem adresu komórki stawiamy symbol dolara przed literą oraz liczbą) przemnóżmy odpowiednie wartości przez kurs.

Zauważmy, że przy adresowaniu bezwzględnym za pomocą dolara blokujemy jednocześnie kolumnę jak i wiersz. Może się zdarzyć, że sytuacja, w której blokować będziemy tylko wiersz lub tylko kolumnę. Wtedy mamy do czynienia z tzw. **adresowaniem mieszanym**. Dobrą ilustracją do tego typu adresowania będzie poniższe zadanie.

## Dodatek Solver

Microsoft Excel oferuje wiele ciekawych dodatków. Jednym z nich jest **Solver**, który pozwala rozwiązywać zagadnienia z tzw. **programowania liniowego**.

Omówmy to zagadnienie na przykładzie.

**Przykład.** Uczniowie z klasy Ia z przyniesionych do szkoły 160 kasztanów, 240 żołędzi i wielu zapałek
maja wykonać ludki dwóch rodzajów. Liczba kasztanów i żołędzi potrzebnych do zbudowania jednego ludka każdego rodzaju przedstawia tabelka:



Obliczyć, ilu ludków kazdego rodzaju powinny zrobic uczniowie, by łaczna liczba ludków była możliwie najwieksza.

Spróbujmy teraz wykonać zadanie z pomocą **Solvera**.

Pierwszym krokiem jest sprawdzenie przez studentów, czy Solver został dodany. Jeśli tak powinien znajdować się w karcie Dane, na samym końcu, w grupie Analiza. Jeśli go tam nie ma, 
należy wybrać Plik->Opcje->Dodatki. Wśród dodatków wybieramy >>Dodatek Solver<<, następnie klikamy przejdź i zaznaczamy odpowiedni kwadracik w okienku, które powinno się pojawić 
i wciskamy Ok. Ikonka powinna pojawić się we wskazanym wyżej miejscu.

Spróbujmy rozwiązać zadanie, które wyżej omówiliśmy. Pytamy jaka jest maksymalna ilość ludków I rodzaju (2 kasztany i 2 żołędzie)
oraz II rodzaju (1 kasztan i 3 żołędzie), jakie możemy wykonać przy użyciu 160 kasztanów i 240 żołędzi. Niech D3 i E3 będą komórkami, które przechowywać będą zmienne x i y odpowiadające 
ilości ludków I i II rodzaju. Opiszmy je literami x i y (w komórkach D2 i E2). W polu G2 wpiszmy suma, którą przechowywać będziemy w komórce G3. W G3 wpiszmy więc formułę: '=D3+E3'. 
Ponieważ komórki D3 i E3 są puste w G3 pojawia się 0. Kliknijmy na G3 i wybierzmy Solvera. Powinno pojawić się okno 'Parametry dodatku Solver'. Jako 'cel' ustawiona powinna być komórka
G3. Jeśli jest inaczej to można na nią kliknąć lub wpisać (adresując bezwzględnie) $G$3. Wskazany 'cel' ustawiamy 'na' 'maks' 'przez zmienianie komórek zmiennych' (wpisujemy lub zaznaczamy
w arkuszu): $D$3:$E$3. W następnym kroku przyciskiem 'Dodaj' będziemy wpisywać kolejne ograniczenia. Przed tym utworzyć musimy dwie komórki, które będą przechowywały ilość zasobów 
(kasztanów i żołędzi) w zależności od liczb x i y. W polach J2 i K2 wpiszmy, odpowiednio 'kasztany' i 'żołędzie', a w komórkach J3 i K3 odpowiednio formuły: '=2*D3+E3', '=2*D3+3*E3'.
Teraz możemy już dodać ograniczenia: $D$3>=0, $E$3>=0, $J$2=<160, $K$3=<240 oraz $D$3:$E$3=int (zaznaczyć komórki i wybrać int/całkowite). Na koniec wybieramy metodę 'LP simpleks' (nieliniowa GRG też da dobry wynik) i wybieramy 'rozwiąż'.
Wyświetlenie rozwiązania poprzedzi wyświetlenie raportu. W polach odpowiednich danych powinny pojawić się wyniki, o ile w wyświetlonym uprzednio oknie raportu taką opcję wybraliśmy. 

**Równania i układy równań.**
